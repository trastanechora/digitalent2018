<p style="text-align: justify">Bagi teman-teman Digitalent, mungkin soal-soal ini terlihat mudah ketika dibaca, namun kenyataanya soai ini sedikit <i>"tricky"</i>, karena import dataframe dari datasets mungkin tidaklah sesederhana ketika import dari .csv file yang kita sudah kerjakan di poin Tugas #1. <i>Let's check it out! </i>:)</p>

<p id="3"><strong><a href="#soal" style="text-decoration: none; color: black;">Jawaban dari soal 3:</a></strong></p>

<p style="text-align: justify">Ok, setelah kita berhasil mengimport iris dari dataset sklearn dan menyimpanya dengan variabel yang bernama "iris", coba kita cek isinya seperti apa dengan cara memanggil namanya:</p>

```python
iris
```

<p style="text-align: justify">Yup! Si iris masih kelihatan <strike>jelek</strike> tidak rapi, bagaimana caranya buat dia rapi seperti di Tugas #1?</p>

<p style="text-align: justify">Perhatikan dataframe yang ingin kita buat, di dalamnya terdapat header sebagai anggota tabel paling atas bukan? sedangkan di bawahnya berisi value yang akan kita hitung entah apanya, dengan kata lain data yang kita butuhin adalah header dan valuenya. Kita lihat iris dari dataset yang berbentuk Dictionary di atas, pertama kita tahu value yang akan kita olah ada di variabel bernama 'data', kemudian kita harus membuat headernya, coba lihat di bagian bawah dari iris, terdapat variabel yang bernama 'feature_names', Yup!</p>

<p style="text-align: justify">Kita tinggal membuat dataframenya dengan bantuan Pandas, <i>But how?</i></p>

<p style="text-align: justify">Ok, kita tentukan dataframe dari iris yang akan dibuat ini kita simpan ke variabel yang kita beri nama 'iris_df', selanjutnya dengan menggunakan method DataFrame() dari Pandas kita bikin dataframenya:</p>

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

<p style="text-align: justify">iris.data adalah value dari variabel yang bernama 'data' di dalam iris, sedangkan iris.feature_names juga kurang lebih seperti itu lah yak.</p>

<p style="text-align: justify">Setelah kita buat berhasil buat dataframenya tanpa notif error, sekarang kita coba lihat hasilnya dengan cara panggil namanya:</p>

```python
iris_df
```

<p style="text-align: justify">Yes! kita telah berhasil membuat Dataframe yang rapi dengan cara import langsung dari sebuah datasets :) <strong><i>Congrats!!</i></strong></p>

<p id="4"><strong><a href="#soal" style="text-decoration: none; color: black;">Jawaban dari soal 4:</a></strong></p>

<p style="text-align: justify">Sebenarnya saya bingung di sini karena soal meminta bikin plot sedangkan cara bikin plot belum diajarkan, tapi ya sudahlah..</p>

```python
x = iris_df.head()

# Simpan 5 baris pertama dari dataframe ke variabel yang kita beri nama x untuk memudahkan pemanggilan di step selanjutnya
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.FacetGrid(x, height=5) \
    .map(plt.scatter, "sepal length (cm)", "sepal width (cm)") \
    .add_legend()

plt.show()
```

<p style="text-align: justify">Di sini kita perlu bantuan Seaborn dan Matplotlib, jiga belum punya mereka maka harus install terlebih dahulu ya.</p>

<p style="text-align: justify">Kita menggunakan plot pola dari FacetGrid dengan metode scatter, "sepal length (cm)" sebagai nilai sumbu X yang dipanggil harus sesuai dengan nama kolom di dataframe, sedangkan "sepal width (cm)" sebagai nilai sumbu Y yang juga harus sama dengan nama kolom di dataframe. Height menyatakan ukuran chartnya.</p>
