<h1>Tugas #1</h1>

Soal:
<ul>
    <li>Data nama rekan anda dan anda (5 orang). Buat distribusi penggunaan huruf pada kelima nama tersebut. Plot pada histogram!</li>
</ul>

<p style="text-align: justify">Pertama-tama kita membuat list yang berisi 5 nama dari teman-teman rekan perjuangan di kelas:</p>

```python
names = ["Maestro Trastanechora", "Yophiza Tulus Ichwansyah", "Muhammad Salma Abdul Aziz", "Dinda Kartika Vilaili", "Nanda Permata Putri"]
```

<p style="text-align: justify">Setelah itu kita coba cek apakah data yang kita buat sudah sesuai dengan apa yang kita inginkan, dengan cara memanggil variabelnya:</p>

```python
names
```

<p style="text-align: justify">OK! Ternyata sudah sesuai..</p>

<p style="text-align: justify">Sekarang kita perlu merubah list nama-nama tersebut menjadi database yang berisi kumpulan huruf. Proses pembuatan database huruf tersebut kurang lebih bisa dibagi menjadi langkah-langkah sederhana seperti berikut:</p>

<ol>
    <li>Menghilangkan sepasi pada nama-nama tersebut, karena spasi bukan terhitung sebagai sebuah karakter/huruf</li>
    <li>Merubah hurufnya menjadi kecil semua atau besar semua, pada kasus ini kita rubah menjadi besar yak. Karena jika tidak maka "M" dan "m" akan terhitung sebagai dua karakter yang berbeda <i>(case sensitifity)</i></li>
    <li>Memecah setiap huruf yang ada pada nama-nama tersebut menjadi list</li>
    <li>Merubah list tersebut menjadi series data supaya bisa dibaca dan diplot menjadi grafik bar</li>
    <li>Membuat grafik bar</li>
</ol>

```python
letterList = []
count = 0

for i in names:
    N = names[count].replace(' ', '').upper()
    M = list(N)
    count += 1
    for J in M:
        letterList.append(J)
```

<p style="text-align: justify">Pertama-tama kita buat variabel yang akan menyimpan list huruf tersebut. Dalam proses ini kita akan menggunakan metode pengulangan "loop for". Untuk setiap list item pada 'names', akan dilakukan penghapusan sepasi, perataan huruf besar dan pemisahan per hurufnya. Variabel 'count' hanya digunakan sebagai penomoran index iterasi saja.</p>

<p style="text-align: justify">"for i in names" artinya jumlah anggota pada list 'names' akan menjadi patokan berapa kali perulangan akan dilakukan, pada kasus ini i yang pertama nilainya 'Maestro Trastanechora', i kedua nilainya 'Yophiza Tulus Ichwansyah', dan seterusnya sampai i yang terakhir yang nilainya 'Nanda Permata Putri'.</p>

<p style="text-align: justify">names[count] berfungsi untuk mengambil nama pada index tertentu dengan variabel count sebagai penentu indexnya yang ditambah 1 setiap kali iterasi berlangsung ("count += 1" adalah singkatan dari "count = count + 1"). names[0] yang berarti 'Maestro Trastanechora', names[1] nilainya 'Yophiza Tulus Ichwansyah', dan seterusnya sampai index terakhir. replace() berfungsi untuk menghapus sepasi, dalam tanda kurung, nilai pertama adalah nilai yang akan diganti dan nilai kedua (setelah koma) adalah nilai penggantinya. upper() berfungsi untuk mengubah semua karakter menjadi upper case. Hasilnya disimpan dalam variabel 'N'. Kemudian nilai 'N' dipisahkan menjadi kumpulan huruf-huruf dengan menggunakan list(), dari 'MAESTROTRASTANECHORA' menjadi  ['M',
 'A',
 'E',
 'S',
 'T',
 'R',
 'O',
 'T',
 'R',
 'A',
 'S',
 'T',
 'A',
 'N',
 'E',
 'C',
 'H',
 'O',
 'R',
 'A']</p>
 
 <p style="text-align: justify">Kemudian di dalam loop kita perlu satu kali lagi melakukan proses loop, yaitu untuk menggabungkan kumpulan huruf dari setiap nama tersebut menjadi utuh dari kelima nama.</p>
 
 <p style="text-align: justify">Note: Ada metode loop yang tanpa menggunakan 'count', yaitu dengan langsung mengubah variabel penghitungnya ke dalam bentuk range. Bentuknya akan menjadi seperti berikut:</p>

```python
# letterList = []

# for i in range(len(names)):
#     N = names[i].replace(' ', '').upper()
#     M = list(N)
#     for J in M:
#         letterList.append(J)
```

<p style="text-align: justify">Setelah itu kita cek apakah hasilnya sudah sesuai dengan harapan, dengan cara memanggil namanya:</p>

```python
letterList
```

<p style="text-align: justify">Yay! Ternyata hasilnya ga PHP. Sekarang kita perlu membuat database yang menampung dan menghitung huruf-huruf tersebut. Kita butuh bantuan Pandas untuk melakukanya. Kita bisa menampungnya sebagai data series kemudian menghitung jumlah huruf yang sama dan menggabungkannya.</p>

```python
import pandas as pd
pd.Series(letterList).value_counts()
```

<p style="text-align: justify">Setelah itu kita cuma tinggal perlu membikin plot grafik bar untuk kepentingan visualisasinya :)</p>

<p style="text-align: justify">Kita akan menggunakan Matplotlib.pyplot untuk menggambarnya. Tapi terlebih dahulu kita simpan dulu data huruf-huruf di atas supaya memudahkan pemanggilannya.</p>

```python
letters = pd.Series(letterList).value_counts()

import matplotlib.pyplot as plt
letters.plot.bar()
plt.show()
```

<h1>Tugas #2</h1>

Soal:
<ul>
    <li>Dari data pada http://rcs.bu.edu/examples/python/data_analysis/Salaries.csv , buat distribusi data yang mungkin ditampilkan (salary, service, dll). Tampilkan dengan plot, scatter, atau histogram (pilih mana yang sesuai)</li>
</ul>

<p style="text-align: justify">Untuk membuka file .csv kita perlu menggunakan pandas, karena kita sudah mengimportnya di atas maka kita tidak perlu import sekali lagi. Langsung saja kita buka filenya dan menampilkannya untuk dianalisis (kita juga bisa menampilkan hanya 5 baris data paling awal sebagai sampel jika mau dengan menggunakan method .head())</p>

```python
dataFrame = pd.read_csv("http://rcs.bu.edu/examples/python/data_analysis/Salaries.csv")
dataFrame
```

<p style="text-align: justify">Selanjutnya kita akan memangkas kolom yang tidak perlu dilibatkan dalam penghitungan kita, dalam kasus ini, saya pribadi akan meniadakan kolom: discipline, phd dan sex. Sehingga menyisakan rank, service dan salary saja. Di sini kita akan membuat 2 grafik: </p>

<ol>
    <li>Menggunakan scatter plot dengan data: rank, service dan salary</li>
    <li>Menggunakan grafik bar dengan data: rank dan rata-rata salary</li>
</ol>

<p style="text-align: justify; font-size: 10px">Sebenarnya di soal hanya disuruh untuk memilih satu saja sih, tapi yasudah lah..</p>

<p style="text-align: justify">Langkah pertama adalah memangkas data yang tidak perlu dari dataframe di atas, kurang lebih seperti ini:</p>

```python
cleanData1 = dataFrame.iloc[:,[0, 3, 5]]
cleanData1.head()
```

<p style="text-align: justify">Command di atas kita menggunakan iloc, di dalam kurung siku ":" berarti semua baris data, sedangkan untuk yang setelah koma adalah nomor index kolom yang mau kita pakai/gunakan/sisihkan.</p>

<p style="text-align: justify">Kita perlu bantuan Seaborn dan matplotlib.pyplot untuk membuat grafiknya. Karena kita sudah pernah import matplotlib di atas ketika mengerjakan tugas nomor satu, maka kita hanya perlu import seaborn saja:</p>

```python
import seaborn as sns

sns.FacetGrid(cleanData1, hue="rank", height=8) \
    .map(plt.scatter, "service", "salary") \
    .add_legend()

plt.show()
```

<p style="text-align: justify">Dari grafik di atas, semakin ke atas maka berarti semakin banyak gaji yang diterima, semakin ke kanan berati semakin senior seseorang telah bekerja yang disimbolkan oleh tahun bekerja. Sedangkan pangkatnya dibedakan oleh warna dari masing-masing titik.</p>

<p style="text-align: justify"><strong>OK!</strong> Sekarang kita akan lanjut membuat grafik yang kedua. Di sini kita akan membuang satu kolom lagi, yaitu kolom 'service'. Kita akan mengelompokkan gaji rata-rata dari tiap rank:</p>

```python
rank1 = cleanData1[ cleanData1['rank'] == 'Prof']
rank2 = cleanData1[ cleanData1['rank'] == 'AssocProf']
rank3 = cleanData1[ cleanData1['rank'] == 'AsstProf']
```

<p style="text-align: justify">Setelah kita berhasil memisahkan ketiga rank tersebut, kita akan mencari rata-rata dari gajinya:</p>

```python
avg1 = rank1['salary'].mean()
avg2 = rank2['salary'].mean()
avg3 = rank3['salary'].mean()
```

<p style="text-align: justify">Setelah itu kita akan menyusun datanya menjadi satu dan membuat grafik bar dari situ:</p>

```python
cleanData2 = {'Prof': avg1, 'AssocProf': avg2, 'AsstProf': avg3}

names = list(cleanData2.keys())
values = list(cleanData2.values())
plt.bar(names, values)

plt.show()
```

<p style="text-align: justify"><strong>Selesai!</strong> Gafik pertama bertujuan untuk menganalisa per satuan titik inividu berdasarkan pangkat, gaji dan masa menjabat. Sedangkan grafik kedua bertujuan untuk menganalisa rata-rata gaji dari masing-masing pangkat.</p>
